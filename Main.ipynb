# Features Selection Part

```python
print("select the facts features and assign they in new data frame movies")
data.columns
movies = data[['id', 'title', 'overview', 'genre']]
movies
```

# Transformation Tasks

The first Task consist to combine the two columns overview and genre to a single column named tags

```python
movies['tags'] = movies['overview']+movies['genre']
movies
```

Now we can drop the two transformed columns *overview* and *genre*

```python
new_movies = movies.drop(columns=['overview', 'genre'])
new_movies
```

# Apply some Algorithmes

1. Bag of word
2. TFIDF

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(new_movies['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

Calculat the distance based on our similarity

```python
#access the similarity for the index number 2
similarity[2]
```

```python
#Convert the similarity of index 2 to list
distance = sorted(list(enumerate(similarity[2])), reverse=True, key = lambda vector:vector[1])
for i in distance[0:5]:
  print(new_movies.iloc[i[0]].title)
```

```python
def recommand(movies):
  index = new_movies[new_movies['title']==movies].index[0]
  distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
  for i in distance[0:5]:
    print(new_movies.iloc[i[0]].title)
```

```python
recommand("Iron Man")
```
